```python
Diwali Sales Analysis  
Unlocking Customer Insights Through Data  

This project performs **Exploratory Data Analysis (EDA)** on Diwali sales data to understand customer demographics, spending behavior, and product preferences.  
Using **Python, Pandas, Matplotlib, and Seaborn**, we clean, process, and visualize the dataset to uncover valuable insights that can help in targeted marketing and business strategies.  
```

```python
Importing Dataset
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Diwali Sales Data.csv", encoding='latin-1')
print(data)
```

```python
Creating Dataframe
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
Data Cleaning
```

```python
df.isna().sum()
```

```python
df.head(6)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1,inplace=True)
print(df.info())
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
#EDA 
```

```python
Sales_by_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending= False)
sns.barplot(x= 'Gender',y= 'Amount',data= Sales_by_gender)

```

```python
From the above charts we can cleary see that majority of the buyers are females and their purchasing power is more than men.
```

```python
ax = sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_by_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending= False)
sns.barplot(x= 'Age Group',y= 'Amount',data=sales_by_age)
```

```python
From the above chart we can see that the majority buyers are from the age group 26-35 and are mostly females.
```

```python
Sales_by_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by= 'Orders',ascending= False)
sns.set(rc={'figure.figsize': (25,3)})
sns.barplot(x= 'State',y='Orders',data=Sales_by_state)
```

```python
Sales_by_state_amt = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending= False)
sns.set(rc={'figure.figsize': (25,3)})
sns.barplot(x= 'State',y='Amount',data=Sales_by_state_amt)
```

```python
From above graphs we understand that most of the orders are from Uttarpradeh, Maharastra and Karnataka.
```

```python
sa = sns.countplot(data=df, x= 'Marital_Status')
sns.set(rc={'figure.figsize':(6,5)})
for bars in sa.containers:
    sa.bar_label(bars)
```

```python
sales_by_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status', y='Amount', hue='Gender', data=sales_by_state)
plt.title("Spending by Marital Status & Gender")
plt.show()
```

```python
From the above graph we understand that most of the buyers are married(women).
```

```python
sa = sns.countplot(data= df,x='Occupation')
sns.set({'figure.figsize' : (25,5)})

for bars in sa.containers:
    sa.bar_label(bars)
```

```python
Sales_by_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by ='Amount', ascending=False)

sns.barplot(data= Sales_by_occupation,x= 'Occupation',y='Amount')
```

```python
From the above charts we can see that majority buyers work in the IT,Helathcare,Aviation sector.
```

```python
sa = sns.countplot(data= df,x='Product_Category')
for bars in sa.containers:
    sa.bar_label(bars)
```

```python
Sales_by_product = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending=False)

sns.barplot(data= Sales_by_product,x='Product_Category',y='Amount')
sns.set({'figure.figsize' : (30,7)})
```

```python
From the above charts we can see that the majority products ordered are food,clothes or electronics.
```

```python
Sales_product = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.barplot(data=Sales_product,x='Product_ID',y='Orders')
```

```python
CONCLUSION
```

```python
From the analysis, we observe that the majority of buyers are married females in the 26–35 age group, primarily from Uttar Pradesh, Maharashtra, and Karnataka, and working in the IT, Healthcare, and Aviation sectors. The most frequently purchased categories were Food, Clothing, and Electronics.
```
